# Map for Data Science

```python
from graphviz import Digraph, nohtml
import os
```

```python
outfname = os.path.join("dot", "datascience")
dot = Digraph(comment='Data science', 
              engine="neato", 
              filename=outfname, format="pdf")
```

```python
nodeids = "A", "B", "C"
names = "knowledge", "mathematics / statistics", "programming"
dirs = "knowledge", "statistics", "programming"
coordinates = "2,3!", "2,0!", "4,2!"
lcoordinates = "1.4,3.5!", "1.8,-0.3!", "4.7,1.7!"
colors = "#fadbd8", "#d6eaf8", "#d5f5e3"
transparency = "5f"; width = "4"; height="4"; i = 0
for nid, name, coord, lcoord, color, directory in zip(nodeids, names, coordinates, lcoordinates, colors, dirs):
    dot.node(name=nid, label="",  
             pos=coord, width=width, height=height,
             labelloc="3,3!",
             shape="circle",  # see shapes, e.g. "component",
             fillcolor=f"{color}{transparency}", style="filled, solid")
    dot.node(name=nid+str(i), 
             label=f"<<b>{name}</b>>", 
             URL="{file}".format(file=os.path.join(os.curdir, directory)),
             # pos="{a},{b}!".format(a=int(coord[0])+1, b=int(coord[2])+1), 
             pos=lcoord, shape="none", fontsize="16")
    i += 1
```
