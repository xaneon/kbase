# Machine Learning

```python
from graphviz import Digraph, nohtml
import os
```

```python
outfname = os.path.join("..", "dot", "machinelearning")
dot = Digraph(comment='Machine learning', engine="neato", filename=outfname, format="pdf")
```

```python
dot.node(name="A", label='<<i> types:</i>>', pos="0,5!", shape="none", fontcolor="lightgrey")  # or shape="note"
dot.node(name="ML", label="machine learning", pos="6,5!", shape="rectangle")

dot.node(name="B", label='<<i> </i>>', pos="0,4!", shape="none", fontcolor="lightgrey")  # or shape="note"
dot.node(name="S", label="supervised", pos="3,4!", shape="rectangle")
dot.node(name="U", label="unsupervised", pos="6,4!", shape="rectangle")
dot.node(name="R", label="reinforcement \nlearning", pos="8,4!", shape="rectangle")

dot.node(name="Cl", label="classification", pos="2,3!", shape="rectangle")
dot.node(name="Re", label="regression", pos="4,3!", shape="rectangle")
dot.node(name="Clu", label="clustering", pos="6,3!", shape="rectangle")


dot.node(name="C", label='<<i> shallow learning:</i>>', pos="0,2!", shape="none", fontcolor="lightgrey")  # or shape="note"
dot.node(name="LogRe", label="logistic regression", pos="2,2!", shape="rectangle")
dot.node(name="LinRe", label="linear regression", pos="4,2!", shape="rectangle")
dot.node(name="kmeans", label="k-means", pos="6,2!", shape="rectangle")

dot.node(name="otherCl", label="Softmax, Perceptron", pos="2,1.5!", shape="none", fontsize="10")
dot.node(name="other", label="kNN, SVM, DecisionTrees, \nRandomForest, Perceptron", pos="3,1.1!", shape="none", fontsize="10")
dot.node(name="otherRe", label="SLR, MLR, \n non-linear regression", pos="4,1.5!", shape="none", fontsize="10")
dot.node(name="otherClu", label="hierachical (e.g. agglomerative), \nDBSCAN, Gaussian mixture", pos="6,1.5!", shape="none", fontsize="10")

dot.node(name="D", label='<<i> deep learning:</i>>', pos="0,0.5!", shape="none", fontcolor="lightgrey")  # or shape="note"
dot.node(name="MLP", label="MLP", pos="2,0.5!", shape="rectangle")
dot.node(name="RBM", label="RBM", pos="6,0.5!", shape="rectangle")

dot.node(name="otherDL", label="DBNs, CNN, RNN, RNTN, LSTM", pos="3,0!", shape="none", fontsize="10")
dot.node(name="autoencoders", label="autoencoders", pos="6,0!", shape="none", fontsize="10")

#TODO

dot.edge("ML", "S", style="solid", arrowhead="none") # rhead="S", arrowhead="vee"
dot.edge("ML", "U", style="solid", arrowhead="none")
dot.edge("ML", "R", style="solid", arrowhead="none")

dot.edge("S", "Cl", style="solid", arrowhead="none")
dot.edge("S", "Re", style="solid", arrowhead="none")
dot.edge("U", "Clu", style="solid", arrowhead="none")

dot.edge("Cl", "LogRe", style="dotted", arrowhead="none")
dot.edge("Re", "LinRe", style="dotted", arrowhead="none")
dot.edge("Clu", "kmeans", style="dotted", arrowhead="none")

#TODO: Deep learning algorithms in nodes
```
